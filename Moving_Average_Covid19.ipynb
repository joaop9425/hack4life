```python
trace = go.Scattergeo(
                     locationmode = 'country names',
                     lon = coordsData['long'],
                     lat = coordsData['lat'],
                     text = coordsData['city_name'],
                     marker = dict(
                            size = data['cases']/5000,
                            color = '#e74c3c',
                            line = {'width': 0.5, 
                                    'color': '#2c3e50'},
                            sizemode = 'area')
                    )
dataTrace = [trace]
layout = go.Layout(
        title = '<b>População americana em 2014</b>',
        titlefont = {'family': 'Arial',
                     'size': 24},
        geo =  {'scope': 'usa',
                'projection': {'type': 'albers usa'},
                'showland': True,
                'landcolor': '#2ecc71',
                'showlakes': True,
                'lakecolor': '#3498db',
                'subunitwidth': 1,
                'subunitcolor': "rgb(255, 255, 255)"
                })
fig = go.Figure(data=dataTrace, layout=layout)
py.iplot(fig)
```
